```python
## Check prime number
def isprime(n):
    for i in range(2,n):
        if n%i == 0:
            return False
    return True
print(isprime(1))
#Optimized : This reduces time complexity from O(n) to O(√n).
def isprime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i == 0:
            return False
    return True
print(isprime(4))

```

```python
### Count word frequency in a sentence
# sent = "Hi hello hello Hi"
# words = sent.split(sep=' ')
# def count_word_frequency_in(sentence):
#     freq_dict={}
#     for word in sentence.split(sep=' '):
#         freq_dict.setdefault(word,sentence.count(word))
#     return freq_dict
# print(count_word_frequency_in(sent))
### Optimal approach (single pass, O(n)):
def count_word_frequency_in(sentence):
    freq = {}
    for word in sentence.split():
        freq[word] = freq.get(word,0)+1
    return freq
sent = "Hi hello hello Hi hello Hello"
print(count_word_frequency_in(sent))
```

```python
### Find second-largest number in list
# def second_largest(int_list):
#     int_list.sort(reverse=True)
#     return(int_list[1])
# print(second_largest([1,4,3,2,7,6]))

### Better approach avoids sorting and runs in O(n):
def second_largest(int_list):
    first=second=float('-inf')
    for n in int_list:
        if n>first:
            second = first
            first = n
        elif first >n > second:
            second =n
    return second
print(second_largest([1,4,3,2,7,7,6]))
```

```python
### Reverse words in a sentence
# def rev_words(sentence):
#     words=sentence.split()
#     words.reverse()
#     print(' '.join(words))
# rev_words("One two three")

### A more Pythonic and efficient way uses slicing:
def rev_words(sentence):
    return ' '.join(sentence.split()[::-1])
rev_words("One two three")
```

```python
# ### Remove duplicates from list (without set)
# def remove_dups(ip_list):
#     op_list = []
#     for i in ip_list:
#         if i not in op_list:
#             op_list.append(i)
#     return op_list
# print(remove_dups([1,1,2,2,3,3,4]))

### Optimal (still without set) uses a dictionary for O(n) lookup:  
def remove_dups(ip_list):
    return list(dict.fromkeys(ip_list))
print(remove_dups([1,1,2,2,3,3,4,1,1,2,2,3,3,4,1,1,2,2,3,3,4,1,1,2,2,3,3,4]))
```

```python
# ### Merge two dictionaries and sum values if keys repeat
# def merge_dicts(dict1,dict2):
#     for i in dict2.keys():
#         if i in dict1.keys():
#             dict1[i] += dict2[i]
#     return dict1
# d1={"a":1,"b":2}
# d2={"a":1,"b":2}
# print(merge_dicts(d1,d2))

### Most optimal approach (O(n))
def merge_dicts(dict1,dict2):
    for k,v in dict2.items():
        dict1[k]= dict1.get(k,0)+v
    return dict1
d1={"a":1,"b":5}
d2={"a":1,"c":2}
print(merge_dicts(d1,d2))
```

Loop drills:

```python
### Nested loop pattern printing (3 patterns)
### Right-angled triangle
for i in range(1,6):
    for j in range(i):
        print("*",end="")
    print()

### Square pattern
for i in range(5):
    for j in range(5):
        print("-",end="")
    print()

### Inverted triangle
for i in range (5,0,-1):
    for j in range(i):
        print("+",end="")
    print()

### Pyramid pattern
n=5
for i in range(1,n+1):       # i = 1 2 3 4 5
    for s in range(n-i):     # s = 4 3 2 1 0
        print(" ",end="")    
    for j in range(2*i-1):   # j = 1 3 5 7 9
        print("*",end="")
    print()

```

```python
### Hollow square
n = 5
for i in range(n):                                       # i = 0        1       2,3    4 
    for j in range(n):                                   # j = 01234    01234   01234  01234
        if i==0 or i == n-1 or j == 0 or j == n-1:       #     *****    *   *   *   *  ***** 
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
### Floyd’s triangle
n = 1
for i in range(1,5):
    for j in range(i):
        print(n,end="")
        n+=1
    print()
```

```python
### Given marks dict → print grade (A/B/C/Fail)
def print_grades(marks_dict):
    for name,marks in marks_dict.items():
        if marks >=50:
            grade = 'A'
        elif marks <=50:
            grade = "F"
        print(name,grade)
marks_dict = {
    "Kavee": 92,
    "Sriman": 78,
    "Mohinth": 65,
    "Anu": 48,
    "Karti": 88
}
print_grades(marks_dict)
```
