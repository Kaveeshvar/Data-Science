```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
# Countrywise Sales
df.groupby(df['Country'])['UnitPrice'].sum()
```

```python
# Country that gave most sales
df.groupby(df['Country'])['UnitPrice'].sum().sort_values(ascending=False).head(1).reset_index()
```

```python
# Country that gave Least sales
df.groupby(df['Country'])['UnitPrice'].sum().sort_values(ascending=True).head(1).reset_index()
```

```python
# Top 15 customers by Purchases
df.groupby(df['CustomerID'])['UnitPrice'].sum().sort_values(ascending=False).head(15).reset_index()
```

```python
df.groupby(['CustomerID', 'Country'])['UnitPrice'] \
  .sum() \
  .sort_values(ascending=False) \
  .head(15) \
  .reset_index()
```

```python
# Country wise highest total spend
df.groupby(['Country', 'CustomerID'])['UnitPrice'].sum() \
  .reset_index() \
  .sort_values(['Country', 'UnitPrice'], ascending=[True, False]) \
  .groupby('Country') \
  .head(1)
```

```python
# Country wise most purchases
df.groupby(['Country', 'CustomerID']) \
  .size() \
  .reset_index(name='Purchases') \
  .sort_values(['Country', 'Purchases'], ascending=[True, False]) \
  .groupby('Country') \
  .head(1)
```
