##### Write 6 short functions (10–15 lines max each):

```python
# Safe divide (handle zero)
def divide(x,y):
    try:
        return x/y
    except ZeroDivisionError: 
        return 'Zero Division error'
print(divide(0,0))
```

```python
# Frequency counter using .get()
def count(ip):
    freq={}
    for i in ip:
        freq[i] = freq.get(i,0)+1
    return freq
print(count([1,2,3,2,1,2]))

# The gold standard of Pythonic
from collections import Counter

Counter([1,2,3,2,1,2])
```

```python
# Normalize a numeric list
def norm(ip):
    mn,mx = min(ip),max(ip)
    return [(x-mn)/(mx-mn) for x in ip]
print(norm([1,2,3,2,1,2]))
```

```python
# Bucket values into low/med/high
def bucket(n):
    if n<=5:
        return 'low'
    elif n<10:
        return 'Medium'
    return 'High'
print(bucket(11))
```

```python
# Clean a text column (strip/lower/replace)
def clean(ip):
    return ip.strip().lower().replace('hi','HI')
print(clean('HI HELLO hi hello HellO'))
```

```python
# Return top N items from a dict by value
def top(ip,n):
    sorted_ip = {x:y for x,y in sorted(ip.items(),key=lambda item:item[1],reverse=True)}
    op = list(sorted_ip.items())[:n]
    return dict(op)
print(top({'apple': 2, 'banana': 3, 'watermelon': 1, 'cherry': 7},2))
```
