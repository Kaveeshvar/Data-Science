Create 2 derived columns:

    CTR = clicks / impressions

    CPC = cost / clicks

```python
df['ctr'] = df['clicks']/df['impressions']
# df['ctr'] = np.where(
#     df['impressions'] > 0,
#     df['clicks']/df['impressions'],0
# )
df['ctr'] = df['ctr'].round(4)
df.head()
```

```python
df['cpc'] = df['cost']/df['clicks']
df['cpc'] = df['cpc'].round(2)
df.head()
```

Answer 3 BUSINESS questions (code + output):

1. Which campaign is most efficient and why?

2. Which campaign is wasting money?

3. What would you recommend to increase ROI?

```python
df
```

```python
# Which campaign is most efficient and why?
q1 = df.groupby('campaign',as_index=False).agg(
    avg_ctr=('ctr','mean'),
    avg_cpc=('cpc','mean')
).sort_values(by=['avg_ctr','avg_cpc'],ascending=[False,True]).head(1)
print(q1,'\n Festive_Offer campaign is best because, it has the highest ctr and lowest cpc')
```

```python
# Which campaign is wasting money?
q2=df.groupby('campaign', as_index=False).agg(
    avg_ctr = ('ctr','mean'),
    avg_cpc = ('cpc','mean')
).sort_values(by=['avg_cpc','avg_ctr'],ascending=[False,True]).head(1)
print(q2, '\nBrand_Awareness is wasting more money because it has very low CTR and high Cpc. -> More spending and less clicks per impressions')
```

```python
# What would you recommend to increase ROI?
q3 = df.groupby('campaign',as_index=False).agg(
    avg_ctr=('ctr','mean'),
    avg_cpc=('cpc','mean')
).sort_values(by=['avg_ctr','avg_cpc'],ascending=[False,True]).head(2)
print(q3,'\nTo improve ROI, I will suggest to increase the spending on Festive_offer and app_launch campaigns. Lower the spending on other two campaigns')
```
