#  Write 6 functions (concise):


```python

# Safe divide (zero-safe)
def safe_divide(num, denom):
    return num / denom if denom != 0 else 0
print(safe_divide(10, 2))  # Output: 5.0
print(safe_divide(10, 0))  # Output: 0
# Frequency counter using .get()
def frequency_counter(items):
    freq = {}
    for item in items:
        freq[item] = freq.get(item, 0) + 1
    return freq
print(frequency_counter(['apple', 'banana', 'apple']))  # Output: {'apple': 2, 'banana': 1}
# Normalize list (min-max)
def normalize_list(lst):
    if not lst:
        return []
    min_val = min(lst)
    max_val = max(lst)
    range_val = max_val - min_val
    return [(x - min_val) / range_val if range_val != 0 else 0 for x in lst]
print(normalize_list([10, 20, 30]))  # Output: [0.0, 0.5, 1.0]
# Top-N from dict by value
def top_n_by_value(d, n):
    return sorted(d.items(), key=lambda item: item[1], reverse=True)[:n]
print(top_n_by_value({'a': 5, 'b': 2, 'c': 8}, 2))  # Output: [('c', 8), ('a', 5)]
# Clean string list (strip/lower/replace)
def clean_string_list(str_list):
    return [s.strip().lower().replace(' ', '_') for s in str_list]
print(clean_string_list([' Hello World ', 'Python Programming ']))  # Output: ['hello_world', 'python_programming']
# Build dict from list of tuples (aggregate values)
def build_dict_from_tuples(tuples_list):
    result = {}
    for key, value in tuples_list:
        result[key] = result.get(key, 0) + value
    return result
print(build_dict_from_tuples([('a', 1), ('b', 2), ('a', 3)]))  # Output: {'a': 4, 'b': 2}
```

# Comprehensions only (no loops):

```python
# Comprehensions only (no loops):

# Filter dict by value
def filter_dict_by_value(d, threshold):
    return {k: v for k, v in d.items() if v > threshold}
print(filter_dict_by_value({'a': 5, 'b': 2, 'c': 8}, 4))  # Output: {'a': 5, 'c': 8}
# Create {campaign: ctr} from two lists
def create_campaign_ctr(campaigns, clicks, impressions):
    return {campaign: safe_divide(clicks[i], impressions[i]) for i, campaign in enumerate(campaigns)}
print(create_campaign_ctr(['camp1', 'camp2'], [100, 50], [1000, 0]))  # Output: {'camp1': 0.1, 'camp2': 0}
```
