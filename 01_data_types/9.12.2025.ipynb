```python
#Assignment 3
import statistics
import numpy as np

def calculate_summary_statistics(data_list):
    """
    Calculates summary statistics for a given list of numerical data.

    Args:
        data_list (list): A list of numbers (e.g., sales figures, times).

    Returns:
        dict: A dictionary containing key descriptive statistics.
    """
    if not data_list:
        return "Error: Data list cannot be empty."

    # Sort data for min, max, median, and quartiles
    data_list_sorted = sorted(data_list)
    n = len(data_list)

    # Calculate statistics
    mean_val = statistics.mean(data_list)
    median_val = statistics.median(data_list)
    
    # --- COMPATIBILITY FIX FOR PYTHON < 3.8 ---
    try:
        # statistics.mode() raises StatisticsError if no unique mode exists
        mode_val = statistics.mode(data_list)
        mode_result = mode_val
    except statistics.StatisticsError:
        # Check if all elements are unique manually for the 'No mode' message
        if len(set(data_list)) == n:
            mode_result = "No single mode (all values unique)"
        else:
            # Handle multimodal data where mode() fails but multimodal() would have worked
            mode_result = "Multimodal data (multiple modes found)"
    # ----------------------------------------

    min_val = min(data_list)
    max_val = max(data_list)
    range_val = max_val - min_val
    
    variance_val = np.var(data_list, ddof=1)
    std_dev_val = np.std(data_list, ddof=1)

    q1_val = np.quantile(data_list, 0.25)
    q3_val = np.quantile(data_list, 0.75)

    # Compile results into a dictionary
    summary_stats = {
        "Mean": mean_val,
        "Median": median_val,
        "Mode": mode_result,
        "Minimum": min_val,
        "Maximum": max_val,
        "Range": range_val,
        "Variance (Sample)": variance_val,
        "Standard Deviation (Sample)": std_dev_val,
        "Q1 (25th Percentile)": q1_val,
        "Q3 (75th Percentile)": q3_val
    }

    return summary_stats

# --- Example Usage (Using data from Assignment 1) ---
sales_data = [120, 125, 128, 130, 135, 137, 138, 140, 142, 145]
stats_output = calculate_summary_statistics(sales_data)

import json
print(json.dumps(stats_output, indent=4))
```
